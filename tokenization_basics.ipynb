```python
import re
# re - regular expression
text = " Hey, It's Arif. "
```

TOKENIZATION

```python
def notebook_tokenize(text):
  parts = re.split(r'([,.:;?_!\"()\']|--|\s)', text)
  tokens = [ p for p in (t.strip() for t in parts) if p != " " ]
  return tokens

tokens = notebook_tokenize(text)
print(tokens)
```

TOKENS --> TOKEN IDs

```python
unique_tokens = sorted(set(tokens))
print(unique_tokens)

special_tokens = ["<|endoftext|>", "<|unk|>"]
vocab_list = unique_tokens + special_tokens
```

```python
# CREATE MAPPINS TOKEN -> ID, ID -> TOKENS

token_to_id = { tok:idx for idx, tok in enumerate(vocab_list)}
id_to_token = { idx:tok for tok, idx in token_to_id.items()}

```

```python
# ENCODE THE TOKEN SEQUENCE

unk_id = token_to_id["<|unk|>"]
encoded_ids = [token_to_id.get(t, unk_id) for t in tokens]
```

```python
print("RAW TEXT:")
print(repr(text))
print("\nTOKENS (after notebook-style tokenizer):")
print(tokens)
print("\nVOCAB LIST (sorted unique tokens + special tokens):")
print(vocab_list)
print("\nTOKEN -> ID mapping:")
print(token_to_id)
print("\nENCODED IDS (sequence):")
print(encoded_ids)
```
